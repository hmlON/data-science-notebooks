# Splitting PDF

```python
from PyPDF2 import PdfFileWriter, PdfFileReader
filename = 'Химмельблау Д. - Прикладное нелинейное программирование.pdf'

inputpdf = PdfFileReader(open(filename, "rb"))
output = PdfFileWriter()

# Set pages count to small number for algoirthm testing
# pages_count = 10

pages_count = inputpdf.getNumPages()

for i in range(pages_count):
    output.addPage(inputpdf.getPage(i))

with open(f"test-{filename}-{pages_count}.pdf", "wb") as outputStream:
    output.write(outputStream)
```

# PDF -> Images list

```python
from PyPDF2 import PdfFileReader
from pdf2image import convert_from_path, convert_from_bytes

filename = f"test-{filename}-{pages_count}.pdf"
pdf_source = open(filename, 'rb')
pdf = PdfFileReader(pdf_source)

images = convert_from_path(filename)
images
```

# Image -> Text (image recognotion)

```python
from pytesseract import image_to_string

texts = [image_to_string(image, lang='rus') for image in images]
for text in texts:
    print(text)
```

```python
full_text = '\n pagebreak \n'.join(texts)
tmp_file = open('tmp_text.txt', 'w')
tmp_file.write(full_text)
tmp_file.close()
```

# Text -> Word document

```python
from docx import Document
document = Document()

for text in texts:
    document.add_paragraph(text)
    document.add_page_break()

document.save('final.docx')
```
